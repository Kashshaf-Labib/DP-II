```python
import pandas as pd
import os

# Find the CSV result file
result_files = [f for f in os.listdir('output_results') if f.endswith('.csv')]
if result_files:
    # Read the CSV file
    results_df = pd.read_csv(f'output_results/{result_files[0]}')
    print("Raw CSV Data:")
    print("=" * 50)
    print(results_df)
    print(f"\nColumns: {list(results_df.columns)}")
    print(f"Shape: {results_df.shape}")

    # Classification counters
    total_images = 0
    correctly_classified = 0  # AI images classified as AI (score > 0.5)
    misclassified = 0  # AI images classified as Real (score <= 0.5)
    unparseable = 0

    results = []

    print("\nDetection Results:")
    print("=" * 50)

    for idx, row in results_df.iterrows():
        filename = row['image'] if 'image' in results_df.columns else row.iloc[0]

        # Look for score in different possible columns
        score = None
        if 'spai' in results_df.columns:
            score = row['spai']
        elif len(results_df.columns) > 3:  # Last column is usually the score
            score = row.iloc[-1]

        if score is not None and isinstance(score, (int, float)):
            total_images += 1

            if score > 0.5:
                prediction = "🤖 AI-Generated"
                status = "✅ CORRECT"
                correctly_classified += 1
            else:
                prediction = "📷 Real Photo"
                status = "❌ MISCLASSIFIED"
                misclassified += 1

            confidence = f"{score:.3f}"
            print(f"{filename}")
            print(f"  → {prediction} (score: {confidence}) {status}")
            print()

            results.append({
                'filename': filename,
                'score': score,
                'prediction': prediction,
                'status': status
            })
        else:
            print(f"{filename}: Unable to parse score - {score}")
            unparseable += 1

    # Summary Statistics
    print("\n" + "=" * 50)
    print("CLASSIFICATION SUMMARY")
    print("=" * 50)
    print(f"Total AI-generated images tested: {total_images}")
    print(f"Correctly classified as AI: {correctly_classified}")
    print(f"Misclassified as Real: {misclassified}")

    if total_images > 0:
        accuracy = (correctly_classified / total_images) * 100
        misclassification_rate = (misclassified / total_images) * 100

        print(f"\n📊 Accuracy: {accuracy:.2f}%")
        print(f"📊 Misclassification Rate: {misclassification_rate:.2f}%")

        # Show misclassified images
        if misclassified > 0:
            print(f"\n❌ Misclassified Images ({misclassified}):")
            print("-" * 50)
            for r in results:
                if r['status'] == "❌ MISCLASSIFIED":
                    print(f"  • {r['filename']} (score: {r['score']:.3f})")

    if unparseable > 0:
        print(f"\n⚠️ Unparseable results: {unparseable}")

    # Save detailed results
    if results:
        results_summary_df = pd.DataFrame(results)
        output_path = 'output_results/classification_summary.csv'
        results_summary_df.to_csv(output_path, index=False)
        print(f"\n💾 Detailed results saved to: {output_path}")

else:
    print("No result files found!")

# Also try to read the raw file content
print("\n" + "="*50)
print("Raw file content (first 500 chars):")
if result_files:
    with open(f'output_results/{result_files[0]}', 'r') as f:
        content = f.read()
        print(content[:500])
```
